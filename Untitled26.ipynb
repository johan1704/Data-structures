```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
dataTrans = pd.DataFrame(scaler.fit_transform(data1), columns=data1.columns)
dataTrans.head()
```

```python
dataTrans.isna().sum()
```

```python
from sklearn.impute import SimpleImputer

imputer = SimpleImputer(strategy='mean')
dataR = pd.DataFrame(imputer.fit_transform(dataTrans), columns=dataTrans.columns)
dataR.head()
```

```python
y = dataR['Appliances']
X = dataR.drop('Appliances', axis=1)
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 42)
```

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train, y_train)

# Predicting the Test set results
y_pred = regressor.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error
mae = mean_absolute_error(y_test, y_pred)
print('This is MAE score for Linear Regression : ', round(mae, 3))
```

```python
residual = (y_test - y_pred)
a=(residual)**2
a
```

```python
from sklearn.metrics import mean_squared_error
mse=mean_squared_error(y_test,y_pred)
import math
rmse=math.sqrt(mse)
rmse
```

```python
from sklearn.metrics import r2_score
score=r2_score(y_test,y_pred)
score
```

```python
from sklearn.linear_model import Ridge
clf = Ridge(alpha=0.4)
clf.fit(X_train, y_train)
y_pred2 = clf.predict(X_test)
```
